## Step 1: Upload CSV file

```python
from google.colab import files

uploaded = files.upload()
```

## Step 2: Load the CSV into a mini SQL database

```python
import pandas as pd
import sqlite3

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Create a small database in memory and load our table into it
conn = sqlite3.connect('my_database.db')
df.to_sql('orders', conn, if_exists='replace', index=False)

print("Loaded", len(df), "rows into a table called 'orders'")
```

## Step 3:  (SELECT)


```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

## Step 4: Filter rows (WHERE)




```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
WHERE Product = 'Laptop' AND TotalPrice > 2000;
"""

pd.read_sql_query(query, conn)
```

## Step 5: Sort results (ORDER BY)


```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

## Step 6: Group and count (GROUP BY + COUNT)


```python
query = """
SELECT Product, COUNT(*) AS order_count
FROM orders
GROUP BY Product;
"""

pd.read_sql_query(query, conn)
```

## Step 7: Group and total up (GROUP BY + SUM/AVG)


```python
query = """
SELECT
    Product,
    SUM(TotalPrice) AS total_revenue,
    AVG(TotalPrice) AS average_order_value
FROM orders
GROUP BY Product
ORDER BY total_revenue DESC;
"""

pd.read_sql_query(query, conn)
```

## Step 8: Filter grouped results (HAVING)

`WHERE` filters individual rows. `HAVING` filters *after* grouping — for
example, only show products with more than $150,000 in total revenue.


```python
query = """
SELECT Product, SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY Product
HAVING SUM(TotalPrice) > 150000
ORDER BY total_revenue DESC;
"""

pd.read_sql_query(query, conn)
```
