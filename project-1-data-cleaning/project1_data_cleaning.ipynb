# Project 1: Data Cleaning & Preparation

Cleaning a messy orders dataset — fixing missing
values, removing duplicates, and fixing formatting — step by step.


## Step 1: Upload CSV file

```python
from google.colab import files

uploaded = files.upload()
```

## Step 2: Load the data into a table (DataFrame)

```python
import pandas as pd

# Get the name of the file you just uploaded
filename = list(uploaded.keys())[0]

# Read it into a DataFrame
df = pd.read_csv(filename)

print("Rows:", len(df))
print("Columns:", len(df.columns))
df.head()
```

## Step 3: Look for missing values

```python
# This shows how many blank/missing values are in each column
df.isnull().sum()
```

## Step 4: Fix the missing values


Fills blank `CouponCode` values with `"No Coupon"`.


```python
# Example: fill missing text values in one column
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

# Check it worked — this column should now show 0 missing
df.isnull().sum()
```

## Step 5: Check for duplicate rows

```python
print("Duplicate rows:", df.duplicated().sum())

# Remove them
df = df.drop_duplicates()
print("Rows after removing duplicates:", len(df))
```

## Step 6: Check for duplicate IDs



```python
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())

df = df.drop_duplicates(subset='OrderID', keep='first')
print("Rows after removing duplicate IDs:", len(df))
```

## Step 7: Fix date formatting

```python
# Convert the Date column to a proper date format (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# errors='coerce' turns any date it can't understand into "NaT" (missing)
# so we can find and check them:
print("Dates that couldn't be read:", df['Date'].isna().sum())
```

## Step 8: Clean up text columns

```python
# Remove extra spaces and make capitalization consistent

text_columns = ['Product', 'PaymentMethod', 'OrderStatus']

for col in text_columns:
    df[col] = df[col].str.strip().str.title()

df[text_columns].head()
```

## Step 9: Save your cleaned data

```python
df.to_csv('cleaned_data.csv', index=False)
print("Saved! Rows:", len(df))

# Download it to  computer
from google.colab import files
files.download('cleaned_data.csv')
```
